```python
import numpy as np
import math
import pandas as pd
```

```python
# print(np.random.seed(42))
# print(np.random.random(size = 2))


#input
X = np.array([np.random.random(size = 2)])
print(X,X.shape)

#output
Y = np.array([[0,1]])
print(Y,Y.shape)

#b
b1 = np.array([0, 0])
b2 = np.array([0, 0])

#weights
w1 = np.array([np.random.random(size = 2), np.random.random(size = 2)])
w2 = np.array([np.random.random(size = 2), np.random.random(size = 2)])
print(w1.shape)
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def forward_prop(X,Y,b1,b2,w1,w2):
    z_2 = np.dot(X,w1) + b1
    a_2 = sigmoid(z_2)
    z_3 = np.dot(a_2,w2) + b2
    a_3 = sigmoid(z_3)
    
    
    return a_3,a_2

def back_prop(a_3,a_2,Y,X,b1,b2,w1,w2):
    delta_3 = (a_3 - Y)*(sigmoid(a_3)*(1-sigmoid(a_3)))
    delta_2 = np.dot(w2,delta_3.T)*(sigmoid(a_2)*(1-sigmoid(a_2)))
    
    der_w2 = np.dot(a_2,delta_3.T)
    der_w1 = np.dot(X, delta_2.T)
    
#     J = der_w1 + der_w2
#     J = np.sum(J)
    
    J = 0.5*np.sum((a_3-Y)**2)
    
    return J, der_w1, der_w2

def update_weights(der_w1, der_w2, w1, w2):
    w1 = w1 - (0.01*(der_w1))
    w2 = w2 - (0.01*(der_w2))
    
    return w1,w2
```

```python
n_iterations = 5
a_3,a_2 = forward_prop(X,Y,b1,b2,w1,w2)
# print(out)

J, der_w1, der_w2 = back_prop(a_3, a_2, Y, X, b1, b2, w1, w2)
# print("Cost: ",J)
# print("derivative of w1:",der_w1)
# print("derivative of w2:",der_w2)

w1,w2 = update_weights(der_w1, der_w2, w1, w2)
print("w1: ",w1)
print("w2: ",w2)
```

```python
n_iterations = 10000
J_history = []
for i in range(n_iterations):
    
    a_3,a_2 = forward_prop(X,Y,b1,b2,w1,w2)
    # print(out)

    J, der_w1, der_w2 = back_prop(a_3, a_2, Y, X, b1, b2, w1, w2)
    J_history.append(J)
    # print("Cost: ",J)
    # print("derivative of w1:",der_w1)
    # print("derivative of w2:",der_w2)

    w1,w2 = update_weights(der_w1, der_w2, w1, w2)
    print("w1: ",w1, "w2: ",w2)
#     print("w2: ",w2)
```

```python
iters = np.array(np.arange(n_iterations))

import matplotlib.pyplot as plt
plt.plot(iters,J_history)
plt.title("J (Cost) vs iterations")
plt.xlabel("Number of iterations")
plt.ylabel("J")
plt.show()
```
